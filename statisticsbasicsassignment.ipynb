Question 1: What is the difference between descriptive statistics and inferential statistics? Explain with examples.

Answer :
 i) Descriptive Statistics:
Descriptive statistics is used to summarize and describe data. It does make predictions or conclusions beyond the data given
- It includes measures like mean , median, mode, range, and standard deviation.
- Data is presented using tables, graphs, and charts,

Example:
If you calculate the average marks of students in your class, you are using descriptive statistics. It only describes the data you have.

ii) Inferential Statistics:
Inferential statistics is used to make predictions or draw conclusions about a population based on a sample.
- It used techniques like hypothesis testing, confidence intervals, and probability.
- It helps in decision-making and forecasting.

Example: If you study marks of 50 students and then predict the performance of all students in the school, you are using inferential statistics.

Key difference:

Descriptive Statistics               Inferential Statistics
Describes data                         Makes predictions
Uses full data                         Uses sample data
No generalization                       Generalizes to population
Example: Average marks of a class          Example: Predicting school performance

Question 2: What is sampling in statistics? Explain the differences between random and stratified sampling.

Answers: Sampling is the process of selecting a small group(sample) from a large population to study and analyze. It helps save time, cost, and effort while still giving useful information about the whole population.

Example: Instead of surveying all students in a school, you select 50 students and study them.

Types of Sampling

i) Random Sampling
In random sampling every individual has an equal chance of being selected.
- Selection is done randomly(like the lottery method.)
- it is simple and unbiased.

Example: Picking 10 student names randomly from a box.

ii) Stratified Sampling:
In stratified sampling, the population is divided into groups(strata) based on characteristics(like age, gender, class), and then samples are taken from each group.
- Ensures all groups are represented.
- More accurate than random sampling in many cases.

Example: Dividing students into boys and girls, then selecting some from each group.

Key Differences:

Random Sampling                              Stratified Sampling
Equal chance for all                      Divided into groups first
No grouping needed                        Requires grouping (strata)
Simpler method                           More accurate and structured
May miss some groups                         Represents all groups

Question 3: Define mean, median, and mode. Explain why these measures of central tendency are important.

Answer:
i) Mean(average)
The mean is the sum of all values divided by the total numbers of values
Formula:Mean = (Sum of observations)/(Number of observation)
Example: For 2,4,6 -> mean = (2+4+6)/3 = 4

ii) Median
The median is the middle value when the data is arranged in ascending or descending order.
- If the number of values is odd -> middle value
- If even - average of two middle values
Example: for 1,3,5 - > median = 3

ii) Mode
The mode is the value that appears most frequently is a datasheet.
Example: For 2, 2,3,4 - > Mode = 2


Importance of measures of central tendency:
- They summarize large data into a single value.
- Help is easy comparison between datasets.
- Useful in decision-making(business, education, etc)
- Show the overall trend or central value of data.
- Make data simple and understandable.

Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?

i) Skewness refers to the measure of asymmetry of a data distribution. It shows whether the data is evenly distributed or tilted to one side.
- Zero skewness - > perfectly symmetrical distribution.
- Positive skewness -> tail extends to the right.
- Negative skewness -> tail extends to the left.

ii) Kurtosis

Kurtosis measures the shape and peak(or flatness) of a distribution. It tells how much data is concentrated in the center or tails.
- High kurtosis(Leptokurtic) -> sharp peak, heavy tails.
- Low Kurtosis(Platykurtic) -> flat peak, light tails.
- Normal Kurtosis(Mesokurtic) -> moderate peak(like normal distribution)


A positive skew(right-skewed distribution) means:
The tail is longer on the right side.
Most values are concentrated on the left(lower values).
A few high values(outliers)pull the mean to the right.
mean >median>mode

Example:
Income distribution - Most people earn lower income, but a few earn very high income.




```python
#6. Implement a Python program to compute the mean, median, and mode of a given list of numbers.

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

import statistics

mean = statistics.mean(numbers)
median = statistics.median(numbers)
mode = statistics.mode(numbers)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

```python
#Question 6: Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python:
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

#convert to arrays
x = np.array(list_x)
y = np.array(list_y)

#covariance matrix
cov_matrix = np.cov(x,y)

#extract covariance value
covariance = cov_matrix[0][1]

#correlation coefficient
correlation = np.corrcoef(x,y)[0][1]

print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

```python
#Question 7: Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result:

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

import matplotlib.pyplot as plt
import numpy as np

#given data
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

#create boxplot
plt.boxplot(data)
plt.title("Boxplot of Given Data")
plt.ylabel("Values")
plt.show()

#calculate Q1, Q3, and IQR
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

#Define bounds

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

#Identify outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("Q1:", Q1)
print("Q3", Q3)
print("IQR", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Outliers:", outliers)
```

```python
'''Q9 You are working as a data analyst in an e-commerce company. The
marketing team wants to know if there is a relationship between advertising spend and
daily sales.
● Explain how you would use covariance and correlation to explore this
relationship.
● Write Python code to compute the correlation between the two lists:
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]'''

#Answer: Understaing the relationshop using covariance and correlation
'''To analyze the relationship between advertising spend and daily sales, we can use two statistical measures: covariances and correlation.
1. Covariance:
Covariance measures how two variables change together:
- If covariance is positive, it means that as advertising spend increases, daily sales also tend to increase.
- If covariance is negative, it indicate that as advertising spend increases, daily sales decreases.
- However, covariance dose not provide a standartized value, so it is difficult  to interpret the strength of the relationship.
2. Correlation:
Correlation provides both the directioon and strength of the relationship between two variables.
- The values ranges from -1 to +1.

* +1 -> perfect positive relationship
* 0 -> No relationship
* -1 -> perfect negative relationship

- A value close to +1 indicates a strong positive relationship, meaning higher advertising spend leasds to higher sales

Thus, by calculating covariance and correlation, we can determine whether increasing advertising spend has a meaningful impact on daily sales.'''

import numpy as np

#Given data
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

#calculate correlation coefficent
correlation = np.corrcoef(advertising_spend, daily_sales)
print("Correlation matrix:\n", correlation)

#Extracting only the correlation value
print("Correlation coefficient:", correlation[0,1])

# Conclusion: The correlation coefficient obtained will be close to +1, indicating a strong positive relationship between advertising spend and daily sales. This suggests that increasing advertsing expenditures is likely to increase sales.

```

```python
'''9. Your team has collected customer satisfaction survey data on a scale of
1-10 and wants to understand its distribution before launching a new product.
● Explain which summary statistics and visualizations (e.g. mean, standard
deviation, histogram) you’d use.
● Write Python code to create a histogram using Matplotlib for the survey data:
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]'''

'''Answer: To analyze customer satisfaction survery data(rated from 1 to 10), the following summary statistics and visualization are useful:

1. Mean(average):
The mean gives the overall range average satisfaction score. It helps understand the general customer sentiment toward the product.

2. Median:
The median represents the middle value when data is sorted.It is useful when there are extreme values(outliers), as it gives a better central tendancy in such cases.

3. Standard deviation:
Standard deviation measures how spread out the scores are from the mean.
- A low standard deviation indicates that most customers gave similar ratings.
- A high standard suggests varied opinions among customers.

4. Minimum and Maximum values:
These show the range of the data and help identify the lowest and highest satisfaction levels.

5. Histogram(visualization):
A histogram displays the frequency distribution of scores. It helps to:
- Identify patterns(e.g, most customers rating 7-9)
- Detect skewness(left or right)
- Spot any unusual gaps or clusters

Using these measures together provides a clear understanding of customer satisfaction before launching a new product.'''


import matplotlib.pyplot as plt

#survery data
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

#create hist
plt.hist(survey_scores, bins=7)

#Add labels and title
plt.xlabel("Survery score")
plt.ylabel("Frequency")
plt.title("Customer Satisfaction Survery Distribution")

#show plot
plt.show()

#Conclusion: By combining summar statistics(mean, median, standard deviation) with a histogram, we can clearly understand how customer satisfaction scores are distributed and make informed decisions about the product launch.



```
